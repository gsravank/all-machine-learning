```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rng = np.random.RandomState(0)
```

```python
def plot_hist(values, bins, title="", figsize=(5, 3)):
    fig, ax = plt.subplots(1, 1, figsize=figsize)
    _ = ax.hist(values, bins=bins)
    _ = ax.set_title(title)
    _ = plt.show()
    return

def plot_2d(x_values, y_values, title="", figsize=(5, 3)):
    fig, ax = plt.subplots(1, 1, figsize=figsize)
    _ = ax.plot(x_values, y_values, 'bo', alpha=0.2)
    _ = ax.set_title(title)
    _ = plt.show()
    return   
```

```python
plot_hist(rng.normal(loc=0.0, scale=1.0, size=1000000), 100)
plot_hist(rng.uniform(-5, 5, size=1000000), 100)
```
